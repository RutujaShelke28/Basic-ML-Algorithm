```python
##First ML model to predict whether the given fruit is whether apple or an orrange
from sklearn.tree import DecisionTreeClassifier
```

```python
Features = [[120, 0],
            [150, 0],
            [130, 1],
            [140, 1]]
Labels = ["apple", "apple", "orange", "orange"]
```

```python
model = DecisionTreeClassifier()
```

```python
model = model.fit(Features, Labels)
```

```python
model.predict([[170,0]])
```

```python
from sklearn.tree import plot_tree
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(4,4))
```

```python
plot_tree(model, feature_names=["weight", "colors"], class_names=["apple","orange"], filled = True)
```
