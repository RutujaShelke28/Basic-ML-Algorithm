```python
##First ML model to predict whether the given fruit is whether apple or an orrange
from sklearn.ensemble import RandomForestClassifier
```

```python
Features = [[120, 0],
            [150, 0],
            [130, 1],
            [140, 1]]
Labels = ["apple", "apple", "orange", "orange"]
```

```python
model = RandomForestClassifier()
```

```python
model = model.fit(Features, Labels)
```

```python
Prediction = model.predict([[170,0]])
```

```python
print(f"This fruit is {Prediction[0]}")
```
